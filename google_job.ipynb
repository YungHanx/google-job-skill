```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        skill = pd.read_csv(os.path.join(dirname, filename))

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session
```

```python
print(skill.head())
```

```python
company_distribution = skill['Company'].value_counts()
```

```python
print(company_distribution)
```

```python
title_distribution = skill['Title'].value_counts().head(10)
print(title_distribution)
```

```python
category_distribution = skill['Category'].value_counts()
print(category_distribution)
```

```python
location_distribution = skill['Location'].value_counts().head(10)
print(location_distribution)
```

```python
plt.figure(figsize=(12, 8))
```

```python
plt.subplot(2, 3, 2)
title_distribution.plot(kind='bar', color='lightgreen')
plt.title('Title Distribution')
```

```python
plt.subplot(2, 2, 3)
plt.xlabel("category",fontsize = 8)
plt.xticks(fontsize=6)
category_distribution.plot(kind='bar', color='lightcoral')
plt.title('Category Distribution')
```

```python
plt.subplot(2, 2, 3)
plt.xticks(fontsize=10)
category_distribution.plot(kind='bar', color='lightcoral')
plt.title('Category Distribution')
```

```python
plt.subplot(2, 3, 4)
location_distribution.plot(kind='bar', color='gold')
plt.title('Location Distribution')
```
